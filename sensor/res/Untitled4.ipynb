```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
file = pd.read_csv('train.csv')
```

```python
file.head(200)
```

```python
x_train = file.drop('Activity',axis=1)
y_train = file['Activity']
```

```python
x_train
```

```python
file_test = pd.read_csv('test.csv')
```

```python
x_test = file_test.drop('Activity',axis=1)
```

```python
y_test = file_test['Activity']
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
error_rate =[]
for i in range(1,120):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    pred_i = knn.predict(x_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
error_rate
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
plt.xlabel('K')
plt.title('Error Rate vs k value')
plt.ylabel('Error Rate')
plt.plot(range(1,120),error_rate,color='blue',linestyle='dashed',marker='o',markerfacecolor='red',markersize=10)
```

```python
knn = KNeighborsClassifier(n_neighbors=6)
knn.fit(x_train,y_train)
pred = knn.predict(x_test)
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```
